```python
import pandas as pd
import numpy as np

# Quadrotor
df_rotor = pd.read_csv('../vlm_body_moments_jacobian.csv')
rotor_mx = df_rotor['Mx_roll'].abs().max()
rotor_my = df_rotor['My_pitch'].abs().max()
rotor_mz = df_rotor['Mz_yaw'].abs().max()

# Fixed wing
df_wing = pd.read_csv('../wing_body_moments_gradient.csv')
wing_mx = df_wing['Mx_roll'].abs().max()
wing_my = df_wing['My_pitch'].abs().max()
wing_mz = df_wing['Mz_yaw'].abs().max()

print("Peak moments (kN·m):")
print(f"{'Configuration':<15} {'|Mx|':>10} {'|My|':>10} {'|Mz|':>10}")
print("-" * 45)
print(f"{'Quadrotor':<15} {rotor_mx/1685:>10.1f} {rotor_my/1685:>10.1f} {rotor_mz/1685:>10.1f}")
print(f"{'Fixed wing':<15} {wing_mx/2992.6:>10.1f} {wing_my/2992.6:>10.1f} {wing_mz/2992.6:>10.1f}")
```
